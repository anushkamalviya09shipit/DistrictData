```python
#Assingment2
list_sugar = [89,120,137,189,90,120,135,123,130,109,88,111]
list_month = ['Jan','Feb','March','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
for num1,num2 in zip(list_sugar,list_month):
  if num1 <= 99:
   print(f"The sugar level for the month of {num2} is {num1} mg/dL. This is Normal")
  elif num1 >= 100 and num1 <= 125:
   print(f"The sugar level for the month of {num2} is {num1} mg/dL. This is Prediabetes")
  else:
   print(f"The sugar level for the month of {num2} is {num1} mg/dL. This is Diabetes")
normal = 0
diab = 0
prediab = 0

for sugar in list_sugar:
  if sugar <= 99:
    normal += 1
  elif sugar >= 100 and sugar <= 125:
    prediab += 1
  else:
    diab += 1

print("Count of months having normal range:",normal)
print("Count of months having prediabetes range:",prediab)
print("Count of months having diabetes range:",diab)

Avg = sum(list_sugar)/len(list_sugar)
print('Average of sugar values:',Avg)
```

```python
def cal_bmi(h,w):
  Bmi = w/(h**2)
  return Bmi
weight = [60,87,54,78]
height = [1.9,2.2,1.6,2]
for h,w in zip(height,weight):
  Bmi = cal_bmi(h,w)
  print('The BMI is:',Bmi)
  if Bmi < 18.5:
    print("underweight")
  elif Bmi > 18.5 and Bmi <= 28:
    print("normal weight")
  else:
    print("overweight")
```
